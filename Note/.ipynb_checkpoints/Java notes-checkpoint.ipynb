### 4️⃣查询容器属性

#### **第一类：顺序容器类**

* C++：vector、deque、list

* Java：ArrayList、LinkedList、ArrayDeque

| 查询内容         | C++ STL                                | Java                                |
| ------------ | -------------------------------------- | ----------------------------------- |
| 元素数量         | v.size()                               | list.size()                         |
| 是否为空         | v.empty()                              | list.isEmpty()                      |
| 是否包含某元素      | find(v.begin(), v.end(), x) != v.end() | list.contains(x)                    |
| 首元素          | v.front()                              | list.get(0)                         |
| 尾元素          | v.back()                               | list.get(list.size() - 1)           |
| 访问第 i 个元素    | v[i] 或 v.at(i)                         | list.get(i)                         |
| 查询第一个匹配元素位置  | find(...)                              | list.indexOf(x)                     |
| 查询最后一个匹配元素位置 | 无直接成员函数，需反向查找                          | list.lastIndexOf(x)                 |
| 容器是否有容量概念    | vector/deque 有，list 无                  | ArrayList 有隐式容量，通常不直接查；LinkedList 无 |
| 查询容量         | v.capacity()                           | 无直接通用写法                             |
| 预留容量         | v.reserve(n)                           | new ArrayList<>(n)                  |
| 清空后数量        | v.clear(); v.size()                    | list.clear(); list.size()           |

<br>

#### **第二类：集合类**

* C++：set、unordered_set

* Java：HashSet、TreeSet

| 查询内容           | C++ STL                | Java               |
| -------------- | ---------------------- | ------------------ |
| 元素数量           | s.size()               | set.size()         |
| 是否为空           | s.empty()              | set.isEmpty()      |
| 是否包含某元素        | s.find(x) != s.end()   | set.contains(x)    |
| 某元素是否存在（另一种写法） | s.count(x)             | set.contains(x)    |
| 最小元素           | *s.begin()（有序 set）     | treeSet.first()    |
| 最大元素           | *prev(s.end())（有序 set） | treeSet.last()     |
| 严格大于 x 的第一个元素  | s.upper_bound(x)       | treeSet.higher(x)  |
| 大于等于 x 的第一个元素  | s.lower_bound(x)       | treeSet.ceiling(x) |
| 严格小于 x 的最大元素   | 需手动处理迭代器               | treeSet.lower(x)   |
| 小于等于 x 的最大元素   | 需手动处理迭代器               | treeSet.floor(x)   |

<br>

#### **第三类：映射类**

* C++：map、unordered_map

* Java：TreeMap、HashMap

| 查询内容           | C++ STL                | Java               |

⚠️**注意**

C++ 中 map[key]

如果 key 不存在，会插入一个默认值。

## 🔴**Java类的种类**

### 1️⃣总表

| 类型  | 关键字                | 用途      |
| --- | ------------------ | ------- |
| 普通类 | class              | 普通对象    |
| 抽象类 | abstract class     | 定义模板    |
| 接口  | interface          | 定义能力    |
| 枚举类 | enum               | 有限集合    |
| 记录类 | record             | 不可变数据对象 |
| 密封类 | sealed             | 限制继承    |
| 内部类 | class inside class | 作用域控制   |
| Object类 | 可隐匿 | 所有类的父类  |

### 2️⃣抽象类

> 用于定义**模板类**。

* 规则      

* 可以有普通方法

* 可以有成员变量

* 可以有构造函数

* 不能直接实例化 

**例子**
```Java
abstract class Animal{

    abstract void speak();

}
class Dog extends Animal{

    void speak(){
        System.out.println("wang");
    }

}
```

### 3️⃣接口

> 接口用于定义**能力**。

* 默认方法是 abstract    

* 变量默认 public static final

* 方法默认 public

**例子**
```Java
abstract class Animal{

    abstract void speak();

}
class Bird implements Flyable{

    public void fly(){
        System.out.println("flying");
    }

}

//函数式接口：只含一个抽象方法
@FunctionalInterface
interface Calculator{
    int calc(int a,int b);
}

Calculator c = (a,b) -> a + b;
```

| 特性   | 抽象类     | 接口         |
| ---- | ------- | ---------- |
| 继承   | extends | implements |
| 多继承  | 不支持     | 支持         |
| 成员变量 | 可以      | 常量         |
| 方法实现 | 可以      | Java8后可以   |
| 用途   | 模板类     | 能力描述       |

**对于函数式接口**
| 接口        | 作用    |
| --------- | ----- |
| Function  | 输入→输出 |
| Consumer  | 消费数据  |
| Supplier  | 提供数据  |
| Predicate | 条件判断  |

**Function**
```Java
Function<Integer,Integer> f = x -> x * x;
```

**Consumer**
```Java
Consumer<String> c = s -> System.out.println(s);
```

**Predicate**
```Java
Predicate<Integer> p = x -> x > 10;
```

### 4️⃣枚举类（enum）

> 用于定义**固定常量集合**，且是一种**特殊类**。

* 可以包含： 

* 字段

* 方法

* 构造函数

| 方法        | 作用     |
| --------- | ------ |
| values()  | 获取全部枚举 |
| valueOf() | 根据名字获取 |
| ordinal() | 获取序号   |

* equals重写规范: 
自反性,
对称性,
传递性,
一致性,
非空。

**例子**
```Java
@Override
public boolean equals(Object obj){

    if(this == obj) return true;

    if(obj == null) return false;

    if(getClass() != obj.getClass()) return false;

    Person p = (Person)obj;

    return name.equals(p.name);
}
```

### 9️⃣补充

#### 1.**instanceof**

* 用于判断一个**在运行时检查对象是否属于特定的类型**，主要用于**确保类型安全转换**

* 语法:**对象引用 instanceof 类型(类或接口)**

1. **A. 检查继承关系**

* `instanceof`不仅检查当前类，还会检查整个继承树。如果**对象是该类本身或其任何子类的实例**，结果均为 `true`。

**示例**
```Java
class Animal {}
class Dog extends Animal {}

Animal myDog = new Dog();
System.out.println(myDog instanceof Dog);    // true
System.out.println(myDog instanceof Animal); // true (子类也是父类的实例)
System.out.println(myDog instanceof Object); // true (所有类都继承自Object)
```

2. **B. 接口实现检查**

* 它可以用来**判断一个类是否实现了某个特定的接口。**

**示例**

```Java
if (myDog instanceof Serializable) {
    // 如果 Dog 实现了 Serializable 接口，则为 true
}
```

## 🟠**注解**

### **普通注解（Annotation）**

> 注解用于**为代码附加元信息**

**特点**

* 不影响程序结构

* 可被编译器或运行时读取

* 常用于框架

**常见注解**

| 注解                   | 作用      |
| -------------------- | ------- |
| @Override            | 标记方法重写  |
| @Deprecated          | 标记过期方法  |
| @SuppressWarnings    | 忽略编译警告  |
| @FunctionalInterface | 标记函数式接口 |
| @SafeVarargs         | 避免泛型警告  |

### **元注解（Meta Annotation）**

> 元注解是**用于修饰注解的注解**

| 元注解         | 作用       |
| ----------- | -------- |
| @Target     | 指定注解使用位置 |
| @Retention  | 指定生命周期   |
| @Documented | 是否生成文档   |
| @Inherited  | 是否被继承    |


#### Target

* 示例：**@Target(ElementType.METHOD)**

| 类型          | 说明   |
| ----------- | ---- |
| TYPE        | 类    |
| METHOD      | 方法   |
| FIELD       | 成员变量 |
| PARAMETER   | 参数   |
| CONSTRUCTOR | 构造函数 |

#### Retention

* 示例：**@Retention(RetentionPolicy.RUNTIME)**

| 类型      | 生命周期 |
| ------- | ---- |
| SOURCE  | 编译前  |
| CLASS   | 编译后  |
| RUNTIME | 运行时  |

### **自定义注解**

**例子**

```Java
//定义
@Target(ElementType.METHOD)
@Retention(RetentionPolicy.RUNTIME)
@interface Log{
    String value();
}

@interface MyAnno{

    String name();

    int age();

}

@interface Test{
    String name() default "unknown";
}

//使用
@Log("login")
public void login(){
}

@MyAnno(name="Tom",age=20)
```

### 1️⃣字节流（InputStream / OutputStream）

**字节流用于处理：**

* 图片

* 音频

* 视频

* 二进制文件

#### **1.InputStream**
**常见子类：**
| 类                    | 作用    |
| -------------------- | ----- |
| FileInputStream      | 文件输入流 |
| BufferedInputStream  | 缓冲输入流 |
| ByteArrayInputStream | 内存输入流 |
| ObjectInputStream    | 对象输入  |

**常用方法**
| 方法           | 作用     |
| ------------ | ------ |
| read()       | 读取1字节  |
| read(byte[]) | 读取多个字节 |
| close()      | 关闭流    |

**示例**
```Java
import java.io.FileInputStream;

FileInputStream fis = new FileInputStream("a.txt");

int data;

while((data = fis.read()) != -1){
    System.out.print((char)data);
}

fis.close();
```

#### **2.OutputStream**

**常见子类：**
| 类                     | 作用   |
| --------------------- | ---- |
| FileOutputStream      | 文件输出 |

| 方法            | 作用     |

| write(byte[]) | 写入字节数组 |

| close()       | 关闭流    |


**示例**
```Java
import java.io.FileOutputStream;

FileOutputStream fos = new FileOutputStream("a.txt");

fos.write("hello".getBytes());

fos.close();
```
